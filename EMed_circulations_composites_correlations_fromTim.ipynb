### Composite anomaly maps (for concurrent time steps only)

#### Create the composite

```python
# Load your time series in the study region (3-7 days for hot/dry/compound)
fwi_weekly = xr.open_mfdataset ('path_to_your_time_series_hotdry.nc', combine = 'nested')
```

```python
# Select JJA
fwi_JJA = fwi_weekly['fwinx'].sel (time = fwi_JJA.time.dt.month.isin([6,7,8]))
```

```python
# Obtain the user-defined composite threshold:
# Question to consider: Under the top XX% hot/drought/hot dry, what does the circulation pattern look like?
high_fwi = fwi_JJA['fwinx'].where (fwi_JJA['fwinx'] > np.quantile (fwi_JJA['fwinx'], .95))
# Check whether it makes sense...
print(high_fwi.count(dim='time').values)
```

```python
# Load global z500 data
# Use xarray to stack up the yearly data
delayed_obj = xr.open_mfdataset('circulations/z500/z500_jjas2014*.nc', combine = 'nested', parallel=True)
with ProgressBar():
    z500_path = delayed_obj.compute()
```

```python
# Select JJA
z500_JJA = z500_path['z'].sel (time = z500_path.time.dt.month.isin([6,7,8]))
# Note: for the above code, the whole JJA is considered as a bulk. 
# Further deseasonalization may be useful, by considering individual months
# e.g.
z500_Jun = z500_path['z'].sel (time = z500_path.time.dt.month.isin([6]))
z500_Jul = z500_path['z'].sel (time = z500_path.time.dt.month.isin([7]))
z500_Aug = z500_path['z'].sel (time = z500_path.time.dt.month.isin([8]))
```

```python
# Select the timesteps of z500 where hot/dry/hotdry are above your threshold
highFWI_z500=z500_JJA.sel(time=high_fwi.dropna(dim='time')['time']).mean(dim = 'time')
```

```python
# Calculate long-term mean
allweek_z500 = z500_JJA.mean(dim = 'time')
allweek_z500_Jun = z500_Jun.mean(dim = 'time')
allweek_z500_Jul = z500_Jul.mean(dim = 'time')
allweek_z500_Aug = z500_Aug.mean(dim = 'time')
```

```python
highFWI_z500.to_netcdf('z500_highFWI.nc')
allweek_z500.to_netcdf ('z500_allweek.nc')
allweek_z500_Jun.to_netcdf ('z500_allweek_Jun.nc')
allweek_z500_Jul.to_netcdf ('z500_allweek_Jul.nc')
allweek_z500_Aug.to_netcdf ('z500_allweek_Aug.nc')
```

### Correlation maps (for concurrent and lagged time steps)

```python
# Propagate the time series into 2d lat-lon field
fwi_lon = np.repeat(fwi_JJA.values[None,...],len(z500_JJA.longitude),axis=0)
fwi_latlon = np.repeat (fwi_lon[None,...],len(z500_JJA.latitude), axis=0)

# Obtain the z500 vs. the hot/dry/hotdry time series correlation coefficient in each grid cell
corrvalue = []
corrcoefMatrix = np.zeros([len(z500_JJA.latitude), len(z500_JJA.longitude)])
for x in range(len(z500_JJA.latitude)):
    for y in range(len(z500_JJA.longitude)):
        # For correlations at concurrent time step
        corrvalue = np.corrcoef(z500_JJA.values[:,x,y],fwi_latlon[x,y,:])
        # For lagged correlations with 1 time step
        #corrvalue = np.corrcoef(z500_JJA.values[0:len(z500_JJA.values)-1,x,y],fwi_latlon[x,y,1:len(z500_JJA.values)])
        corrcoefMatrix[x,y] = corrvalue[0,1]

corrcoefMatrix = np.squeeze(np.asarray(corrcoefMatrix))
    
print (np.shape(corrcoefMatrix))

corr = Dataset('z500_vs_FWI.nc', 'w', format='NETCDF4')
corr.description = 'z500_vs_FWI'
corr.createDimension('latitude', len(z500_JJA.latitude))
corr.createDimension('longitude', len(z500_JJA.longitude))
corr.createDimension('z500_vs_FWI', 1)

latitude  = corr.createVariable('latitude', 'f4', ('latitude',))
longitude = corr.createVariable('longitude', 'f4', ('longitude',))
CorrCoef   = corr.createVariable('z500_vs_FWI', 'f4', ('latitude',   'longitude', 'z500_vs_FWI'), fill_value=-999.0)
""
latitude.units = 'degree_north'
longitude.units = 'degree_east'
latitude[:]  = z500_JJA.latitude
longitude[:] = z500_JJA.longitude
CorrCoef[:,:,0]   = corrcoefMatrix[:,:]

corr.close()
```
